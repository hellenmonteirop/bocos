```python
import matplotlib.pyplot as plt

def primos_ate(n):
    primos=list(range(2,n+1))
    compostos=list(range(2,n+1))
    idx=0
    while idx<len(primos):
        p=primos[idx]
        compostos.remove(p)
        j=p+1
        while j<=n:
            if j%p==0:
                if j in primos:
                    primos.remove(j)
            j=j+1
        idx=idx+1
    return primos

def maior_fator_primo_de(n):
    for p in primos_ate(n)[::-1]:
        if n%p==0:
            return p
    

def e_primo(n):
    if (n==0) or (n==1):
         return False
    for k in list(range(2,n,1)):
            if n%k == 0:
                return False
    return True


def fib(n):
    a=1
    b=0
    while (n>0):
        a,b=b,(a+b)
        n-=1
    return b


def lista_fib(n):
    lista=[]
    aux=0
    while aux<n:
        lista.append(fib(aux))
        aux=aux+1
    return lista

def lista_eficiente_fib(n):
    lista=[0,1]
    aux=0
    while aux<(n-2):
        lista.append((lista[-1]+lista[-2]))
        aux=aux+1
    return lista

def fib_ate(n):
    lista=[]
    aux=0
    while fib(aux)<(n+1):
        lista.append(fib(aux))
        aux=aux+1
    return lista



def proporcao_primos(n):
    aux=1
    lista=[]
    while aux<n:
        if e_primo(aux)==True:
            lista.append(aux)
    aux=aux+1
    return len(lista)/n
```

```python
i=2
n=5000
l1=[0,0]
l2=[1,3]
while i<=n:
    l1.append(((len(primos_ate(i)))))
    l2.append(((len(fib_ate(i)))))
    
    
    i=i+1
    
print(i,primos_ate(i),'\n',fib_ate(i))
#print(i,'primos:',primos_ate(i),'freq:',l1,len(l1),len(primos_ate(i)),'\n','  fibo:',fib_ate(i),'freq:',l2,len(l2),len(fib_ate(i-1)),'\n')
```

```python

print(l1,l2)          
plt.plot(range(0,n+1),l1,'-o',label='Números primos')
plt.plot(range(0,n+1),l2,'-x',label='Números da sequência de Fibonacci')
plt.grid()
plt.legend()
plt.title("Número absoluto de elementos por naturais")
plt.show()
```

```python
m=1
l1_norm=[l1[0]]*len(l1)
l2_norm=[l2[0]]*len(l2)
for k in range(1,len(l1)):
     l1_norm[k]=l1[k]/m
     l2_norm[k]=l2[k]/m
     m=m+1          
plt.plot(range(0,n+1),l1_norm,'-o',label='Números primos')
plt.plot(range(0,n+1),l2_norm,'-x',label='Números da sequência de Fibonacci')
plt.grid()
plt.legend()
plt.title("Proporção elementos por naturais")

plt.show()
```
